```python
import scanpy as sc
import pandas as pd
import hdf5plugin
```

```python
df = pd.read_csv("./data/9.Marker_genes/brain_markers.csv")
df
```

```python
astrocyte_marker_lst = [
    "GFAP",
    "FGFR3",
    "SLC4A4",
    "SLC14A1",
    "NFIA",
    "CLU",
    "GABRB1",
    "NTRK2",
    "SPARCL1",
    "SLC1A2",
    "DTNA",
    "GLUL",
    "SLC1A3",
    "PTGDS",
    "RYR3",
    "ATP1A2",
]


microglial_marker_lst = [
    "GRIA2",
    "ATP1A3",
    "PKM",
    "PGAM1",
    "GNAS",
    "PEBP1",
    "ALDOA",
    "KIF1A",
    "SLC1A3",
    "APBB1IP",
    "CD74",
    "DOCK8",
    "KCNMA1",
    "FTL",
    "SORL1",
    "FKBP5"
]


oligodendrocytes_marker_lst = [
    "PLP1",
    "GSN",
    "CLDN11",
    "TF",
    "PTGDS",
    "CNDP1",
    "CNTNAP2",
    "QKI",
    "PCDH9",
    "ZEB2",
    "SKG1",
    "TMEFF2",
    "TCF12",
    "MOBP",
    "MBP",
    "RTN4",
]
```

```python
# Filter the rows for each cell type with their specific genes
astrocyte_filtered = df[(df['cell_name'] == 'Astrocyte') & (df['Symbol'].isin(astrocyte_marker_lst))]
microglial_filtered = df[(df['cell_name'] == 'Microglial cell') & (df['Symbol'].isin(microglial_marker_lst))]
oligodendrocytes_filtered = df[(df['cell_name'] == 'Oligodendrocyte') & (df['Symbol'].isin(oligodendrocytes_marker_lst))]

# Filter the rows for all other cell types
other_cells = df[~df['cell_name'].isin(['Astrocyte', 'Microglial cell', 'Oligodendrocyte'])]

# Combine the filtered rows for the specific cell types with the other cell types rows
filtered_df = pd.concat([astrocyte_filtered, microglial_filtered, oligodendrocytes_filtered, other_cells])
```

```python
filtered_df.to_csv("./data/9.Marker_genes/brain_markers4display.csv", index=False)
```
